### Pandas merge

```python
import pandas as pd
```

```python
pedidos = pd.read_csv("../ficheros_curso/merge/Pedidos.txt", sep=';')
pedidos.head()
```

```python
empresas = pd.read_csv("../ficheros_curso/merge/Empresas.txt", sep=';')
empresas.head()
```

```python
empleados = pd.read_csv("../ficheros_curso/merge/Empleados.txt", sep=';')
empleados.head()
```

```python
ped_empr = pd.merge(pedidos, empresas, left_on="idempresa", right_on="id")
ped_empr.head()
```

```python
final = pd.merge(ped_empr, empleados, left_on="idempleado", right_on="id")
final.head()
```

```python
final.drop(columns=["idempleado","idempresa","id_x","id_y","cargo"], inplace=True)
final.head()
```

```python
final.rename(columns={"nombre":"empleado"}, inplace=True)
final.head()
```

Pruebas con groupby

```python
# Calcular el total de facturación por cada empresa de transporte:
final.importe.groupby(final.empresa).sum()
```
